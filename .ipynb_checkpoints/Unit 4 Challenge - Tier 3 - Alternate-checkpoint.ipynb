```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose
properties = properties.transpose()
properties.head()
```

```python
# Reset index
properties.index
properties = properties.reset_index()
properties.head()
```

```python
# Assign the values of the first row to column headings
properties.columns = properties.iloc[0]
properties = properties.drop(index=0)
properties = properties.reset_index(drop=True)
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename "Unnamed: 0" and "NaT" columns to "Borough" and "Code"
properties = properties.rename(columns={
    properties.columns[0]: 'Borough',
    properties.columns[1]: 'Code'
})
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Convert to long format, all prices on separate rows
properties = properties.melt(
    id_vars=['Borough', 'Code'],
    var_name='Date',
    value_name='Price'
)
properties.head()
```

```python
# Check column names
properties.iloc[:, :5]
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check column data types
properties.dtypes
```

```python
# Auto convert data types
properties = properties.convert_dtypes()
properties.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Check columns for missing and unique values
pd.DataFrame({
    'Not Missing': properties.notna().sum(),
    'Missing': properties.isna().sum(),
    'Unique': properties.nunique()
})
```

```python
# Check unique values for each Borough
properties['Borough'].value_counts()
```

```python
# Check unique values for each Code
properties['Code'].value_counts()
```

```python
# Check amount of missing Code data for each Borough
properties.groupby('Borough')['Code'].agg(lambda x: x.isna().sum())
```

```python
# Check amount of missing Price data for each Borough
properties.groupby('Borough')['Code'].agg(lambda x: x.isna().sum())
```

```python
# List to remove 3 rows with missing data and 9 rows that aren't boroughs
drop_rows = [
'EAST MIDLANDS',
'EAST OF ENGLAND',
'LONDON',
'NORTH WEST',
'NORTH EAST',
'SOUTH EAST',
'SOUTH WEST',
'Unnamed: 34',
'Unnamed: 37',
'Unnamed: 47',
'WEST MIDLANDS',
'YORKS & THE HUMBER'
]
print(drop_rows)
```

```python
# Complete creating valid dataset by removing non-borough rows
properties_valid = properties.drop(properties[properties['Borough'].isin(drop_rows)].index)

# Check unique values for columns in valid dataset
pd.DataFrame({
    'Not Missing': properties_valid.notna().sum(),
    'Missing': properties_valid.isna().sum(),
    'Unique': properties_valid.nunique()
})

# Number of Boroughs is now correct
```

```python
# Compare to unique values in previous datset
pd.DataFrame({
    'Not Missing': properties.notna().sum(),
    'Missing': properties.isna().sum(),
    'Unique': properties.nunique()
})

# Not missing data has been reduced from previous dataset
```

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Alternate strategy - create DataFrame with just 1998 and 2018 prices for each borough
```

```python
# Create Borough list

borough_list = properties_valid['Borough'].unique().tolist()
print(borough_list, len(borough_list))
```

```python
# For loop to create summary DataFrame

# Empty list to hold results
year_summary_list = []

# Iterate through the list of boroughs
for borough in borough_list:

    # Filter for the current borough
    borough_data = properties_valid[properties_valid['Borough'] == borough]
    
    # Compute the annual average prices
    annual_avg = borough_data.groupby('Year')['Price'].mean()

    # Extract the values for 1998 and 2018
    price_1998 = annual_avg.loc[1998]
    price_2018 = annual_avg.loc[2018]
    
    # Append the borough name and ratio to the "ratios" list as a tuple
    year_summary_list.append((borough, price_1998, price_2018))

# Convert the list of tuples to a DataFrame
year_summary_df = pd.DataFrame(year_summary_list, columns=['Borough', 'Average Price 1998', 'Average Price 2018'])

# Display without scientific notation
pd.set_option('display.float_format', '{:,.2f}'.format)
print(year_summary_df)
```

```python
# Add ratio column to DataFrame
year_summary_df['Ratio'] = year_summary_df['Average Price 1998'] / year_summary_df['Average Price 2018']
print(year_summary_df)
```

```python
# Sort DataFrame

year_summary_df.sort_values(by='Ratio', ascending=False, inplace=True)
print(year_summary_df)
```
